````python
# Import necessary modules
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
````

````python
school_path = "data/Schools_LAUSD.geojson"
school = gpd.read_file(school_path)
school
````

###
In the file with the locations of schools the addresses are available however the is no geometry so using geocode addresses will be added to the data frame

````python
parks_path = "data/Recreation_and_Parks.geojson"
park = gpd.read_file(parks_path)
park
````

````python
school = school.to_crs(epsg=3067)
park = park.to_crs(epsg=3067)
````

````python
park['centroid'] = park.centroid
park.head()
````

````python
from shapely.ops import nearest_points
def get_nearest_values(row, other_gdf, point_column, value_column="geometry"):
    """Find the nearest point and return the corresponding value from specified value column."""
    
    # Create an union of the other GeoDataFrame's geometries:
    other_points = other_gdf["centroid"].unary_union
    
    # Find the nearest points
    nearest_geoms = nearest_points(row[point_column], other_points)
    
    # Get corresponding values from the other df
    nearest_data = other_gdf.loc[other_gdf["centroid"] == nearest_geoms[1]]
    
    nearest_value = nearest_data[value_column].values[0]
    
    return nearest_value
````

````python
from shapely.ops import nearest_points
school["nearest_park"] = school.apply(get_nearest_values, other_gdf=park, point_column="geometry",value_column="Name",axis=1)
````

````python
school.head()
````

````python
import folium
from pyproj import crs
import geopandas as gpd
import matplotlib.pyplot as plt
````

````python
# Convert points to GeoJSON

school = school.to_crs(epsg=4326)
school_gjson = folium.features.GeoJson(school, name="Parks Closes to Schools")
school_gjson
````

````python
# Create a Map instance
m = folium.Map(location=[34.0522, -118.2437], zoom_start=12, control_scale=True,tiles='cartodbpositron',
               prefer_canvas=True)

folium.GeoJson(school,  
                name='schools in the LA Unified Schoo School District (LAUSD)',
                style_function=lambda x: {'color':'transparent','fillColor':'transparent','weight':0},
                tooltip=folium.features.GeoJsonTooltip(fields=['FULLNAME','nearest_park'],
                                                        aliases = ['School Name','Nearest Park Name'],
                                                        labels=True,
                                                        sticky=False,
                                                       toLocaleString=True
                                                                    )
               ).add_to(m)
folium.LayerControl().add_to(m)
m
````

````python
outfp = "school_map.html"
m.save(outfp)
````
